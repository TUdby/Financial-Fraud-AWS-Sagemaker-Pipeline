# Data Processing

In this section we wish to get our data into the processed state so we can then ingest it into Feature Store. We face several problems with the data in its current state. First, we have categorical columns that our rcf and xgb model we will run wont allow. However, if we simply dummify the categorical data we are left with far too many features and training times would become lengthy, so we would want to do some form of dimensionality reduction. However, PCA is not meaningful on categorical data. After some research, I found a dimensionality reduction process called Factor Analysis of Mixed Data (FAMD) that creates meaningfull dimensionality reduction for data that has both numerical and categorical features.

Below you can see the data, which is from an IEEE competition on Kaggle that took place three years ago. I don't have the knowledge to do a too intricate job in data engineering so I simply join the two tables provided and perform FAMD.

```python
%%time
import pandas as pd

train_identity = pd.read_csv('./../Data/train_identity.csv') #, nrows=nrows)
train_transaction = pd.read_csv('./../Data/train_transaction.csv') #, nrows=nrows)

data = train_transaction.join(train_identity, on='TransactionID', lsuffix='', rsuffix='-identity')

data.head()
```

## Factor Analysis of Mixed Data (FAMD)


A dimensionality reduction algorithm that works on combinations of categorical and numerical data. Sources:

https://en.wikipedia.org/wiki/Factor_analysis_of_mixed_data

https://towardsdatascience.com/famd-how-to-generalize-pca-to-categorical-and-numerical-data-2ddbeb2b9210

The steps
1. Standard scale numerical variables (get z-score)
2. For categorical
    - One-hot encode
    - Divide each column by root of its probability
    - center columns
3. Apply PCA


Due to the very large dataset the serial processing time for FAMD becomes excessive. To ease this, I have set up a parallel version of my original code. 

## As a Local Job

First we seperate the labels from the features and drop columns that have too many nan values.

```python
label = data['isFraud']
data = data.drop(['isFraud', 'TransactionID'], axis=1)

data.dropna(thresh=data.shape[0]*0.15, axis=1, inplace=True)
```

```python
data.shape
```

Then I identity which columns are numerical and which are categorical.

```python
# get list of numerical vs categorical columns
numericals = data.select_dtypes(include='number').columns
categoricals = list(set(data.columns)-set(numericals))
print(numericals)
print(categoricals)
```

The following is a script that the multiple processes will use on each column. One handles scaling the numericals and the other scales the categoricals.

```python
%%writefile scale.py
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import math

scaler = StandardScaler()

# These functions will be passed a single column to be processed by its own process

# this function takes a numerical column, imputes the mean into empty areas, and standard scales it
def scale_nums(series):
    series = series.fillna(series.mean())
    return pd.Series(np.transpose(scaler.fit_transform(pd.DataFrame(series)))[0])

# scale_cats takes a categorical dummy column and divides by the sqrt of the probability of the class
def scale_cats(series):
    series /= math.sqrt((sum(series) / len(series)))
    return series
```

Here we perform the first two steps of FAMD

```python
%%time
from multiprocessing import Pool, cpu_count
import scale

if __name__=='__main__':
    
    # get as many processes as is helpful. due to the large size of our 
    # dataset this should be bounded by the amount of cores
    num_processes = min(data.shape[1], cpu_count())
    
    # this first pool deals with numerical columns. We put them in a list and
    # map them to their own processes to be standard scaled. Then they are
    # reassembled and placed back into our dataframe
    with Pool(num_processes) as pool:

        seq = [data[col_name] for col_name in numericals]
        results = pool.map(scale.scale_nums, seq)

        data[numericals] = pd.concat(results, axis=1)
    
    # before moving to the categorical processing, we need to dummify our categories
    data = pd.get_dummies(data)
    
    # this second pool takes the categorical dummy columns in a list and maps them to
    # processes to be scaled by their probability. They are then reassembled and 
    # placed back into our dataframe.
    with Pool(num_processes) as pool:

        cats = list(set(data.columns)-set(numericals))
        seq = [data[col_name] for col_name in cats]
        
        results = pool.map(scale.scale_cats, seq)
        
        data[cats] = pd.concat(results, axis=1)
        
    # Nan values in categorical columns are treated as their own class
    # by pd.get_dummys, and for the numericals we imputed the mean into
    # nan values. However, for any numerical column that was completely 
    # nan values, nothing was done and so it must be dropped before we do PCA
    data = data.drop(data.columns[data.isna().any()], axis=1)


data.head()
```

And for the third step we conduct pca reduction.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = min(32, data.shape[1]))
data = pca.fit_transform(data.values)
```

```python
data.shape
```

We re-label the data and that finishes up preprocessing (the label was seperated at the beginning so that the labels didn't go through FAMD).

```python
data = pd.concat(
    [
        label,
        pd.DataFrame(data)
    ],
    axis=1
)
```

# Feature Store

Now that we have processed our data, we can ingest it into feature store. Note that this is not necessary for simply training a model, but for professional ML production it is good as it creates a shared, reliable storage of features over time.

## As a Local Job

First, we must append and event time column and a unique id column.

```python
import time

cur_time = int(round(time.time()))
data['EventTime'] = pd.Series([cur_time] * len(data), dtype='float64')
data['RecordID'] = pd.Series([i for i in range(len(data))], dtype='int')

data.head()
```

Then we define a feature group. Recall that a feature group essentially defines the table that will store the features.

```python
from sagemaker import Session

boto_session = boto3.Session(region_name = region)
sagemaker_client = boto_session.client('sagemaker', region)
fs_runtime = boto_session.client('sagemaker-featurestore-runtime', region)

featurestore_session = Session(
    boto_session=boto_session,
    sagemaker_client=sagemaker_client,
    sagemaker_featurestore_runtime_client=fs_runtime
)
```

```python
from sagemaker.feature_store.feature_group import FeatureGroup
from time import gmtime, strftime

fg_name = f'bank-fraud-pca-reduced-{strftime("%d-%H-%M-%S", gmtime())}'
fg = FeatureGroup(
    name = fg_name,
    sagemaker_session=featurestore_session
)

fg.load_feature_definitions(data_frame=data)
```

Now that we have defined it, we can create the feature group and ingest the data.

```python
prefix='bank-fraud-featurestore'
s3_uri = 's3://{}/bank_fraud/{}/'.format(bucket, prefix)
fg.create(
    s3_uri= s3_uri,
    record_identifier_name='RecordID',
    event_time_feature_name='EventTime',
    role_arn=execution_role,
    enable_online_store=True
)

while fg.describe().get('FeatureGroupStatus') != 'Created':
    print('.', end='')
    time.sleep(5)
    
print('!')
```

```python
fg.ingest(
    data_frame=data, 
    max_processes=3, # processes created with workers to work in parallel
    max_workers=3, # threads within each process to work in parallel
    wait=True
)
```

Here I confirm that the data is ingested (at least on the online store).

```python
fs_runtime.get_record(
    FeatureGroupName=fg_name,
    RecordIdentifierValueAsString=str(100)
)
```

That took 45 mins! The processing job on cloud instances will be faster.

## Random Cut Forest (RCF)

Define Data Channels

```python
from sagemaker import TrainingInput

train_channel = TrainingInput(
    s3_data=rcf_path,
    content_type='text/csv;label_size=0',
    distribution='ShardedByS3Key'
)

rcf_data = {'train':train_channel}
```

Train

```python
from sagemaker.estimator import Estimator
from sagemaker.image_uris import retrieve

components = 32

container = retrieve('RandomCutForest', region)
prefix = 'rcf_bank_fraud'

rcf_estimator = Estimator(
    image_uri=container,
    role=execution_role,
    instance_count=1,
    instance_type='ml.m5.large',
    output_path='s3://{}/{}/output'.format(bucket, prefix)
)

rcf_estimator.set_hyperparameters(feature_dim=components)
```

```python
rcf_estimator.fit(rcf_data)
```

Now that we have trained our RCF model, lets deploy it and send samples of our validation data for prediction.

```python
rcf_predictor = rcf_estimator.deploy(
    endpoint_name='rcf-fraud',
    initial_instance_count=1,
    instance_type='ml.m5.large'
)

rcf_predictor.serializer = sagemaker.serializers.CSVSerializer()
rcf_predictor.deserializer = sagemaker.deserializers.CSVDeserializer()
```

```python
sample = val.to_csv(index=False, header=False).split(sep='\r\n')
sample[0:2]
```

```python
scores = []
batch_size = 10000
n=0
while (n+1)*batch_size < len(sample):
    scores += rcf_predictor.predict(sample[n*batch_size:(n+1)*batch_size])
    n += 1
    
scores += rcf_predictor.predict(sample[n*batch_size:-1])
scores = [float(score[0]) for score in scores]
```

```python
rcf_predictor.delete_endpoint()
```

RCF gives anomoly scores. Below, I create lists identifying data as anomalous based on how many standard deviations from the mean its score is.

```python
from numpy import mean, std

mean = mean(scores)
std = std(scores)

bound_1std = mean + std
bound_2std = mean + 2 * std
bound_3std = mean + 3 * std

predictions_1std = [int(is_outlier) for is_outlier in scores > bound_1std]
predictions_2std = [int(is_outlier) for is_outlier in scores > bound_2std]
predictions_3std = [int(is_outlier) for is_outlier in scores > bound_3std]
```

With this data, lets plot confusion matrices to see how these threre different criteria (1std, 2std, and 2std from the mean) predict

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

ConfusionMatrixDisplay(
    confusion_matrix(
        y_true=labels.values, 
        y_pred=predictions_3std
    )
).plot()

print('Accuracy:', accuracy_score(y_true=labels.values, y_pred=predictions_3std))
print('Recall:', recall_score(y_true=labels.values, y_pred=predictions_3std))
print('Precision:', precision_score(y_true=labels.values, y_pred=predictions_3std))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(
        y_true=labels.values, 
        y_pred=predictions_2std)
).plot()

print('Accuracy:', accuracy_score(y_true=labels.values, y_pred=predictions_2std))
print('Recall:', recall_score(y_true=labels.values, y_pred=predictions_2std))
print('Precision:', precision_score(y_true=labels.values, y_pred=predictions_2std))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(
        y_true=labels.values, 
        y_pred=predictions_1std)
).plot()

print('Accuracy:', accuracy_score(y_true=labels.values, y_pred=predictions_1std))
print('Recall:', recall_score(y_true=labels.values, y_pred=predictions_1std))
print('Precision:', precision_score(y_true=labels.values, y_pred=predictions_1std))
```

Notice that as we decrease the std from the mean that we consider anomolous, we decrease accuracy and precsion but increase recall.

Remember:
Accuracy is the probability that we are correct in our prediction: 

$$ P( P_t \cap P_p \cup N_t \cap N_p ) $$

Recall is the probability that we will predict positive given that the label is truely postitive:

$$ P(P_p | P_t ) $$

Precision is the probability that a label is positive (1) given that we predict it to be so:

$$ P(P_t | P_p) $$

## Extreme Gradient Boost (XGB)

Define data channels

```python
training_data_channel = TrainingInput(
    s3_data=xgb_path,
    content_type='text/csv'
)

validation_data_channel = TrainingInput(
    s3_data=val_path,
    content_type='text/csv'
)
```

```python
from sagemaker import image_uris
from sagemaker.estimator import Estimator

container = image_uris.retrieve('xgboost', region, version='latest')

xgb_estimator = Estimator(
    image_uri=container,
    role=execution_role,
    instance_count=1,
    instance_type='ml.m5.xlarge',
    output_path='s3://{}/{}/output'.format(bucket, prefix)
)
    
xgb_estimator.set_hyperparameters(
    objective='binary:logistic',
    num_round=200,
    early_stopping_rounds=10
)
```

```python
xgb_estimator.fit({
    'train':training_data_channel,
    'validation':validation_data_channel
})
```

```python
xgb_predictor = xgb_estimator.deploy(
    endpoint_name='xgb-fraud',
    initial_instance_count=1,
    instance_type='ml.t2.medium'
)
xgb_predictor.serializer = sagemaker.serializers.CSVSerializer()
xgb_predictor.deserializer = sagemaker.deserializers.CSVDeserializer()
```

```python
sample = val.to_csv(index=False, header=False)\
                .split(sep='\r\n')
sample[0:2]
```

```python
predictions = []
batch_size = 10000
n=0
while (n+1)*batch_size < len(sample):
    predictions += xgb_predictor.predict(sample[n*batch_size:(n+1)*batch_size])[0]
    n += 1
    
predictions += xgb_predictor.predict(sample[n*batch_size:-1])[0]
predictions = [ round(float(predict)) for predict in predictions ]
```

```python
xgb_predictor.delete_endpoint()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, recall_score, precision_score

ConfusionMatrixDisplay(
    confusion_matrix(
        y_true=labels.values, 
        y_pred=predictions
    )
).plot()

print('Accuracy:', accuracy_score(y_true=labels.values, y_pred=predictions))
print('Recall:', recall_score(y_true=labels.values, y_pred=predictions))
print('Precision:', precision_score(y_true=labels.values, y_pred=predictions))
```
